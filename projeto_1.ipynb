```python
candidatos = []
aprovados = []

def cadastro():  
    prox = input(str("\nDeseja cadastrar um candidato?\n0 - NÃO\n1 - SIM\nOPÇÃO: "))
    while (prox == "1"): 
        candidato = []
        candidato.append(input("Digite o nome do candidato(a):"))
        eX = input("Nota da entrevista: ")
        candidato.append(eX)
        tX = input("Nota do teste teorico:")
        candidato.append(tX)
        pX = input("Nota da teste pratico:")
        candidato.append(pX)
        sX = input("Nota da Avaliação soft ")
        candidato.append(sX)

        candidatos.append(candidato)
        prox = input(str("\nDeseja cadastrar um novo candidato?\n0 - NÃO: \n1 - SIM: "))

def buscaCandidato():
    print("\n###  NOTAS MÍNIMAS PARA APROVAÇÃO:  ###")
    eNota = input("Digite média do entrevista:")
    tNota = input("Digite média do teorico:")
    pNota = input("Digite média do pratico:")
    sNota = input("Digite média do soft:")


    cVar = True
    for i in range(len(candidatos)):
        candNotas = candidatos[i]
        aprovadosPrint = ("")
        if int(candNotas[1])>=int(eNota) and int(candNotas[2])>=int(tNota) and int(candNotas[3])>=int(pNota) and int(candNotas[4])>=int(sNota):
            aprovadosPrint = "Candidato: {}\nNotas: e{}_t{}_p{}_s{}\n".format(candNotas[0],candNotas[1],candNotas[2],candNotas[3],candNotas[4])
            if cVar:
              print("\n###  LISTA DE APROVADOS  ###\n")
              cVar = False

            print(aprovadosPrint)
        if aprovadosPrint == (""):
          print("\n### NENHUM APROVADO ###\n")
1
print("###  CADASTRO DE CANDIDATOS  ###\n")
cadastro()
buscaCandidato()
```
